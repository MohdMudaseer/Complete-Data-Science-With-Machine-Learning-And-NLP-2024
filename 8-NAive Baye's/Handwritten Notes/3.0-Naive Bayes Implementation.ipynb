## Naive Baye's Machine Learning Algorithms Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
```

```python
df1=sns.load_dataset('tips')
df1
```

```python
from sklearn.preprocessing import OneHotEncoder

enc=OneHotEncoder()
encoded=enc.fit_transform(df1[['day','sex']]).toarray()
```

```python
encoder_df=pd.DataFrame(encoded,columns=enc.get_feature_names_out())
```

```python
encoder_df
```

```python
df1=pd.concat([df1,encoder_df],axis=1)
```

```python
df1
```

```python
df1.drop(columns=['day', 'sex'], inplace=True)
```

```python
df1
```

```python
df1['smoker'].value_counts()
```

```python
df1['time'].value_counts()
```

```python
# transforming time
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
df1['time']=enc.fit_transform(df1['time'])
```

```python
df1.head()
```

```python
df1['smoker']=df1['smoker'].map({'Yes': 1, 'No': 0})
```

```python
X=df1.drop('smoker',axis=1)
```

```python
X
```

```python
y=df1['smoker']
```

```python
y
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gau=GaussianNB()
gau.fit(X_train,y_train)
```

```python
y_pred=gau.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
param_grid = [
    {
        'solver': ['liblinear'],
        'penalty': ['l1', 'l2'],
        'C': [0.01, 0.1, 1, 10],
        'class_weight': [None, 'balanced']
    },
    {
        'solver': ['lbfgs', 'sag', 'newton-cg'],
        'penalty': ['l2'],
        'C': [0.01, 0.1, 1, 10],
        'class_weight': [None, 'balanced']
    },
    {
        'solver': ['saga'],
        'penalty': ['l1', 'l2'],
        'C': [0.01, 0.1, 1, 10],
        'class_weight': [None, 'balanced']
        # ❌ No l1_ratio here
    },
    {
        'solver': ['saga'],
        'penalty': ['elasticnet'],
        'C': [0.01, 0.1, 1, 10],
        'l1_ratio': [0.1, 0.5, 0.9],  # ✅ Only used for elasticnet
        'class_weight': [None, 'balanced']
    }
]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV,StratifiedKFold

reg=LogisticRegression(max_iter=7000)
grid=GridSearchCV(estimator=reg,param_grid=param_grid,scoring='accuracy',refit=True,cv=StratifiedKFold())
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
```
